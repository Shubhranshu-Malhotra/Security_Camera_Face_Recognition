## Method Details:
1. Use `Deep Learning` res10 model for **face detection**.
2. Use `face_recognition.face_encoding()` for **encoding faces**.
3. Use `SVM` model for **face recognition**.

```python
from imutils import paths
import numpy as np
import argparse
import imutils
import pickle
import cv2
import os
import wget
import face_recognition
import matplotlib.pyplot as plt
print("Imports Successful !!")
```

```python
DETECT_CONF_THRESH = 0.5
RECOG_CONF_THRESH = 0.5
image_folder = "D:/Projects/security_camera_face_recognition/Security_Camera_Face_Recognition/face_recognition/datasets/caltech_faces_train/"
detector_main_folder = "D:/Projects/security_camera_face_recognition/Security_Camera_Face_Recognition/face_recognition/DL_dlib_SVM/face_detection_model"
encodings_save_path = "D:/Projects/security_camera_face_recognition/Security_Camera_Face_Recognition/face_recognition/DL_dlib_SVM/output/caltech_train_encodings_m2.pickle"

# Extra
embedding_model_path = "D:/Projects/security_camera_face_recognition/Security_Camera_Face_Recognition/face_recognition/DL_dlib_SVM/nn4.small2.v1.t7"
```

```python
# Load face detector model
print("[INFO] loading face detector...")
protoPath = os.path.join(detector_main_folder, "deploy.prototxt")
modelPath = os.path.join(detector_main_folder,"res10_300x300_ssd_iter_140000.caffemodel")
detector = cv2.dnn.readNetFromCaffe(protoPath, modelPath)
print("[INFO] face detector loaded.")
```

```python
print("[INFO] loading face recognizer...")
embedder = cv2.dnn.readNetFromTorch(embedding_model_path)
```

```python
# grab the paths to the input images in our dataset
print("[INFO] loading images' paths...")
imagePaths = list(paths.list_images(image_folder))
print("Sample image paths:")
print(imagePaths[:10])
print("Total images loaded: ",len(imagePaths))
knownEncodings = []
knownNames = []
# initialize the total number of faces processed
total = 0
```

```python
# loop over the image paths
for (i, imagePath) in enumerate(imagePaths):
    # extract the person name from the image path
    print("[INFO] processing image {}/{}".format(i + 1, len(imagePaths)))
    name = imagePath.split('/')[-1].split(os.path.sep)[-2]
    image = cv2.imread(imagePath)
    image = imutils.resize(image, width=600)
    (h, w) = image.shape[:2]

    # construct a blob from the image
    imageBlob = cv2.dnn.blobFromImage( cv2.resize(image, (300, 300)), 1.0, (300, 300),
                                        (104.0, 177.0, 123.0), swapRB=False, crop=False)
    # apply OpenCV's deep learning-based face detector to localize faces in the input image
    detector.setInput(imageBlob)
    detections = detector.forward()

    # ensure at least one face was found
    if len(detections) > 0:
        # we're making the assumption that each image has only ONE face, so find the bounding box with the largest probability
        i = np.argmax(detections[0, 0, :, 2])
        confidence = detections[0, 0, i, 2]
        
        if confidence > DETECT_CONF_THRESH:
            # compute the (x, y)-coordinates of the bounding box for the face
            box = detections[0, 0, i, 3:7] * np.array([w, h, w, h])
#             print(box)
            (startX, startY, endX, endY) = box.astype("int")
#             print("original coordinates (top left, bottom right): ", (startX, startY, endX, endY))
            # Need to convert because the face_recognition.face_encoding() doesn't use coordinate system
            box = [(startY, endX, endY, startX)]
#             print("Converted coordinates: ", box)
#             detected_face = image.copy()
#             cv2.rectangle(detected_face, (startX, startY), (endX, endY),(0, 0, 255), 2)
#             plt.imshow(detected_face)
#             box = [tuple(box.astype('int'))]
            
#             (startX, startY, endX, endY) = box.astype("int")
#             print(box)
            # extract the face ROI and grab the ROI dimensions
#             face = image[startY:endY, startX:endX]
#             (fH, fW) = face.shape[:2]
#             # ensure the face width and height are sufficiently large
#             if fW < 20 or fH < 20:
#                 continue

            # construct a blob for the face ROI, then pass the blob
            # through our face embedding model to obtain the 128-d
            # quantification of the face
#             faceBlob = cv2.dnn.blobFromImage(image, 1.0 / 255,
#                 (96, 96), (0, 0, 0), swapRB=True, crop=False)
            rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
#             detected_face = rgb.copy()
#             cv2.rectangle(detected_face, (startX, startY), (endX, endY),(0, 0, 255), 2)
#             plt.imshow(detected_face)
#             break
            # Compute encoding using face_recognition library
            encodings = np.array(face_recognition.face_encodings(rgb, box))
            encodings = encodings.flatten()
#             print(type(encodings), encodings.shape, encodings)
#             print(type(encodings.flatten()), encodings.flatten().shape, encodings.flatten())
    
#             embedder.setInput(faceBlob)
#             vec = embedder.forward()
#             print(type(vec), vec.shape, vec)
#             print(type(vec.flatten()), vec.flatten().shape, vec.flatten())
            
            # add the name of the person + corresponding face
            # embedding to their respective lists
            knownNames.append(name)
            knownEncodings.append(encodings)
            total += 1
    print("[INFO] processed image {}/{}".format(total, len(imagePaths)))

# dump the facial embeddings + names to disk
print("[INFO] serializing {} encodings...".format(total))
data = {"encodings": knownEncodings, "names": knownNames}
f = open(encodings_save_path, "wb")
f.write(pickle.dumps(data))
f.close()
```

```python
# loop over the image paths
for (i, imagePath) in enumerate(imagePaths):
    # extract the person name from the image path
    print("[INFO] processing image {}/{}".format(i + 1, len(imagePaths)))
    name = imagePath.split('/')[-1].split(os.path.sep)[-2]
    image = cv2.imread(imagePath)
    image = imutils.resize(image, width=600)
    (h, w) = image.shape[:2]

    # construct a blob from the image
    imageBlob = cv2.dnn.blobFromImage( cv2.resize(image, (300, 300)), 1.0, (300, 300),
                                        (104.0, 177.0, 123.0), swapRB=False, crop=False)
    # apply OpenCV's deep learning-based face detector to localize faces in the input image
    detector.setInput(imageBlob)
    detections = detector.forward()

    # ensure at least one face was found
    if len(detections) > 0:
        # we're making the assumption that each image has only ONE face, 
        # so find the bounding box with the largest probability
        i = np.argmax(detections[0, 0, :, 2])
        confidence = detections[0, 0, i, 2]
        
        if confidence > DETECT_CONF_THRESH:
            # compute the (x, y)-coordinates of the bounding box for the face
            box = detections[0, 0, i, 3:7] * np.array([w, h, w, h])
            (startX, startY, endX, endY) = box.astype("int")

            # Need to convert because the face_recognition.face_encoding() doesn't use coordinate system
            box = [(startY, endX, endY, startX)]
            rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

            # Compute encoding using face_recognition library
            encodings = np.array(face_recognition.face_encodings(rgb, box))
            encodings = encodings.flatten()

            # add the name of the person + corresponding face
            # embedding to their respective lists
            knownNames.append(name)
            knownEncodings.append(encodings)
            total += 1
    print("[INFO] processed image {}/{}".format(total, len(imagePaths)))

# dump the facial embeddings + names to disk
print("[INFO] serializing {} encodings...".format(total))
data = {"encodings": knownEncodings, "names": knownNames}
f = open(encodings_save_path, "wb")
f.write(pickle.dumps(data))
f.close()
```

```python
knownEncodings = []
knownNames = []

# loop over the image paths
for (i, imagePath) in enumerate(imagePaths):
    # extract the person name from the image path
    print("[INFO] processing image {}/{}".format(i + 1,
        len(imagePaths)))
    name = imagePath.split('/')[-1].split(os.path.sep)[-2]
    # load the input image and convert it from BGR (OpenCV ordering)
    # to dlib ordering (RGB)
    image = cv2.imread(imagePath)
    rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # detect the (x, y)-coordinates of the bounding boxes
    # corresponding to each face in the input image
    boxes = face_recognition.face_locations(rgb,model = 'hog')

    # compute the facial embedding for the face
    encodings = np.array(face_recognition.face_encodings(rgb, boxes))
#     print(type(encodings), encodings.shape, encodings)
    # loop over the encodings
    for encoding in encodings:
        # add each encoding + name to our set of known names and
        # encodings
        knownEncodings.append(encoding)
        knownNames.append(name)
        if(len(encodings) > 1 or len(boxes) > 1 ):
            print(len(encodings), len(boxes))
            print(type(encodings), encodings.shape, encodings)
            detected_face = image.copy()
            for box in boxes:
                (startX, startY, endX, endY) = box
                cv2.rectangle(detected_face, (startY, endX), (endY, startX),(0, 0, 255), 2)
            plt.figure()
            plt.imshow(detected_face)

# dump the facial embeddings + names to disk
print("[INFO] serializing {} encodings...".format(len(knownEncodings)))
data = {"encodings": knownEncodings, "names": knownNames}
f = open(encodings_save_path, "wb")
f.write(pickle.dumps(data))
f.close()
```
